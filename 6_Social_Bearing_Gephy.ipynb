```python
import pandas as pd
pd.set_option('display.max_colwidth', 0)
```

```python
df = pd.read_csv('tweets.csv')
df
```

```python
df2 = df.loc[:][['Handle','Tweet']]
df2
```

```python
def handler(tweet):
    list = []
    check = "@"
    for x in tweet.split():
        if check in x:
            x = x.replace(":","")
            x = x.replace("@","")
            x = x.replace(".","")
            x = x.replace("!","")
            list.append(x)
    return list
```

```python
df2['test'] = df.Tweet.apply(lambda x: handler(x))
```

```python
df2[~df2.Tweet.str.contains("@") == False]
```

```python
df3 = pd.DataFrame(columns=['source', 'target'])
```

```python
for row in df2.itertuples(index=False):
    for x in row.test:
        #print(row.date, tag)
        df3 = df3.append({'source': row.Handle, 'target': x}, ignore_index=True)
```

```python
df3['target'] = df3['target'].str.lower()
df3['source'] = df3['source'].str.lower()
df3
```
